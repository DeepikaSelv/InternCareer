```python
correlation_matrix = df[['job_level','salary']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
plt.hist(df['rating'])
plt.xlabel('rating')
plt.ylabel('awards')
plt.title('Awards Rating Distribution')
plt.show()
```

```python
plt.figure(figsize=(10, 6))  
sns.boxplot( x="age", y='awards', data=df, )
plt.show()
```

```python
columns_to_plot = ['Dept'] 

plt.figure(figsize=(5,8))
for i, column in enumerate(columns_to_plot, start=1):
    plt.subplot(1, len(columns_to_plot), i)
    counts = df[column].value_counts().sort_index()
    plt.pie(counts, labels=counts.index, autopct='%1.1f%%', startangle=360)
    plt.title(f'{column} Ratings')
plt.tight_layout()
plt.show()
```
